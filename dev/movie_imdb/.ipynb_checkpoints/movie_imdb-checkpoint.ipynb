```python
set(list(df_full["genres"]))
```

```python
s_genre_split = df_full["genres"].map(lambda x: x.replace(" ","").split("|"))
s_genre_flaten = [item for sublist in list(s_genre_split) for item in sublist]
l_genre = list(set(s_genre_flaten))
```

```python
for genre in l_genre:
    genre_cleaned = genre.replace("-","_")
    df_full["fe_genre_" + genre_cleaned] = s_genre_split.map(lambda x: 1 if genre in x else 0)
```

```python
col_to_keep = [col for col in df_full.columns if col.startswith("fe_")]
df_full[[
    "genre",
] + col_to_keep]
```
